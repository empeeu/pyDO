Very well, so the goal now is to efficiently calculate the following for an arbitrary function:
\begin{align}
E\left[\mathcal L(\mathbf u) \gamma_i\right] \\
\left<\mathcal L(\mathbf u), \mathbf u_j \right> 
\end{align}
Substitute the decomposition
\begin{align}
E\left[\mathcal L(\mathbf u_k \gamma_k) \gamma_i\right]\\
\left<\mathcal L(\mathbf u_k \gamma_k), \mathbf u_j \right> 
\end{align}

\begin{align}
E\left[\mathcal L\left(\mathbf u_k(\mathbf x, t) \gamma_k(t ; \xi)\right) \gamma_i(t; \xi)\right] 
&\approx E\left[\mathcal L\left(\mathbf u_k(\mathbf x, t)\right) \gamma_k(t ; \xi) \gamma_i(t; \xi)\right] \\
&= \mathcal L\left(\mathbf u_k(\mathbf x, t)\right) E\left[\gamma_k(t ; \xi) \gamma_i(t; \xi)\right] \\
&= \mathcal L\left(\mathbf u_k \right) E\left[\gamma_k \gamma_i\right] 
\end{align}

Similarly:
\begin{align}
\left<\mathcal L(\mathbf u_k \gamma_k), \mathbf u_j \right>  
&\approx \left<\mathcal L(\mathbf u_k) \gamma_k, \mathbf u_j \right>  \\
&= \left<\mathcal L(\mathbf u_k), \mathbf u_j \right> \gamma_k
\end{align}

# A new example

For this example, I'm just going to create some dummy data, and then I'm going to test how well we are approximating the quantities above using this nodal approach (and we can compare it to the real, expensive thing)

```python
# Pick a function
L = lambda u: np.sin(u) # u**2 #+ np.abs(u) + 
```

```python
def eig_sort(C):
    w, v = np.linalg.eig(C)
    I = np.argsort(w)[::-1]
    w = w[I]
    v = v[:, I]
    return w, v
def eval_and_plot():
    expectation_n = np.dot(L(un), np.dot(yn.T, yo_i) / N_MC)
    inner_product_n = np.dot(yn, np.dot(L(un).T, uo_i))
    plot(x, expectation, '--')
    gca().set_color_cycle(None)
    plot(x, expectation_n)

    figure()
    for i in range(N_DO):
        hist(inner_product_n[:, i], 40, histtype='step')
    gca().set_color_cycle(None)
    for i in range(N_DO):
        hist(inner_product[:, i], 40, histtype='step', linewidth=2)
def plot_nodes(nodes=None):
    figure()
    subplot(5, 5, 1)
    count = 1
    for i in range(n_do):
        for j in range(n_do):
            subplot(5, 5, count)
            count += 1
            if i != j:
                plot(yo_i[:, j], yo_i[:, i], '.')
                if nodes!=None:
                    plot(nodes[:, j], nodes[:, i], 'xr')
            else:
                hist(yo_i[:, i], 40, histtype='step')
                if nodes!=None:
                    plot(nodes[:, i], [20] * nodes[:, i].size, 'or')
            title (str(i) + ' -- ' + str(j))
```

```python
N_MC = 1000
N_X = 100
N_DO = 5
x = np.linspace(-1, 1, N_X)[:, None]
rvs = np.random.randn(5, N_MC)
u = np.sin(rvs[0, :] * x * np.pi) * np.cos(rvs[1, :] * x * np.pi) + x**2 * rvs[2, :] ** 2 * rvs[3, :] + rvs[4, :] * x**3 + rvs[0, :]
plot(x, u[:, :15])
```

```python
# Put into DO format
C_uu = np.dot(u, u.T)
w, v = eig_sort(C_uu)
w = np.real(w)
v = np.real(v)
n_do = N_DO 
u_i = v[:, :n_do] # Modal DO modes
y_i = np.dot((u).T, u_i) # Modal DO coefficients
# Rotate New modal coefficients (rotated so uncorrelated)
C = np.dot(y_i.T, y_i) / (N_MC - 1) 
w, v = eig_sort(C)
yo_i = np.dot(y_i, v) / np.sqrt(w) #New yo_i's
#np.dot(yo_i.T, yo_i) / (n_mc - 1) # Verify
uo_i = np.dot(u_i, v) * np.sqrt(w) #New uo_i's
#u1 = np.dot(u_i, y_i.T) # Verify
#u2 = np.dot(uo_i, yo_i.T) # Verify
#np.abs(u1 - u2).sum() # Verify
plot(x, uo_i)
figure()
#for i in range(n_do):
#    hist(yo_i[:, n_do - 1 -i], 40)
plot_nodes()
```

```python
# reconstruct the realizations (and compare?)
u_r = np.dot(uo_i, yo_i.T)
print u_r.shape, u.shape, np.sqrt(((u-u_r)**2).sum()) / u.shape[0], np.abs(u-u_r).max(), np.abs(u-u_r).min()
plot(x, u[:, -5:], '-k')
plot(x, u[:, -5:], '--')
```

Recall:
\begin{align}
E\left[\mathcal L\left(\mathbf u_k(\mathbf x, t) \gamma_k(t ; \xi)\right) \gamma_i(t; \xi)\right] 
&\approx E\left[\mathcal L\left(\mathbf u_k(\mathbf x, t)\right) \gamma_k(t ; \xi) \gamma_i(t; \xi)\right] \\
&= \mathcal L\left(\mathbf u_k(\mathbf x, t)\right) E\left[\gamma_k(t ; \xi) \gamma_i(t; \xi)\right] \\
&= \mathcal L\left(\mathbf u_k \right) E\left[\gamma_k \gamma_i\right] 
\end{align}

Similarly:
\begin{align}
\left<\mathcal L(\mathbf u_k \gamma_k), \mathbf u_j \right>  
&\approx \left<\mathcal L(\mathbf u_k) \gamma_k, \mathbf u_j \right>  \\
&= \left<\mathcal L(\mathbf u_k), \mathbf u_j \right> \gamma_k
\end{align}

```python
# Now we pick some random nodes, and we calculate the same quantities
n_nodes = 5
ids = np.round(np.random.rand(n_nodes) * N_MC).astype(int)
```

We want:
$$ \gamma_M A = \mathbf I = \gamma_N$$
$$A = (\gamma_M)^{-1}$$
But in this case $\gamma_M \in R^{n_{nodes}\times n_{DO}}$, and $n_{nodes} \ge n_{DO}$, so we have to take the pseudo-inverse of a non-square matrix. 
Once we know $A$, we can multiple all the coefficients by $A$. And as usual, we also have to modify the modes. Since
$$u_r = u_M \cdot \gamma_M^T$$
Then
$$u_r = (u \cdot \gamma_M) \cdot (\gamma_M \gamma_M^{-1})^T$$
$$u_r = u_N \cdot \gamma_M^T$$


```python
A = np.linalg.pinv(yo_i[ids, :])
yn = np.dot(yo_i, A)
un = np.dot(uo_i, yo_i[ids, :].T)
# Quick check
((np.dot(un, yn.T) - u_r) ** 2).sum()
# Evaluate the quantities approximately and plot
eval_and_plot()
plot_nodes(yo_i[ids, :])
```

###Try 2
So apparently picking random nodes doesn't work well at all. Let's see if we can pick some good nodes based on the covariance matrix?

```python
yo_m = np.mean(yo_i, 0)
C = np.dot((yo_i - yo_m).T, (yo_i - yo_m)) / N_MC
C2 = np.dot((yo_i).T, (yo_i))
U, S, V = np.linalg.svd(C)
#find the max in each column
ids = []
yt = yo_i - yo_m
for i in range(n_nodes):
    #Project nodes unto this basis
    p = np.dot(yt, U)
    ids.append(np.argmax(np.abs(p).sum(1)))
    # remove this nodal basis from the samples
    b = np.dot(yt, yt[ids[-1], :])
    d = yt[ids[-1], :]
    c = np.dot(d, d)
    a = np.dot((b / c)[:, None], d[None, :])
    yt = yt - a
ids
```

```python
#pick those nodes and plot
A = np.linalg.pinv(yo_i[ids, :])
yn = np.dot(yo_i, A)
un = np.dot(uo_i, yo_i[ids, :].T)
yn[ids, :]
# Evaluate the quantities approximately
eval_and_plot()
plot_nodes(yo_i[ids, :])
```
